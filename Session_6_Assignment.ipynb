### Titanic Dataset (df2)

```python
path2 = '../session_6/train_and_test2.csv'

df2 = pd.read_csv(path2)
df2.info()
```

```python
df2.head()
```

#### Exercise 1: Calculate Gender-Based Survival Percentage

```python
# count of gender based on survival:
gender_survival = df2.groupby('Sex')['survived'].sum()
gender_survival
```

```python
total_pax_by_gender = df2['Sex'].value_counts()
total_pax_by_gender
```

- First, we create a Dataframe below that shows gender based survival and total passengers

- Then we concatenate 1 column that shows percentage of survival by gender

- From above we can see gender also has already been one-hot encoded into integers, looking at the lower survival absolute value of '0', we infer that '0' is Male and '1' is Female. I will then map the dataframe to an accurate gender naming as the index

```python
gender_df = pd.DataFrame({
    'Gender Survival': gender_survival,
    'Total Passengers': total_pax_by_gender
})
```

```python
gender_labels = {0: 'Male', 1: 'Female'}
gender_df.index = gender_df.index.map(gender_labels)
gender_df
```

```python
gender_df['Survival Percentage'] = ((gender_df['Gender Survival'] / gender_df['Total Passengers'])*100).round(2)

gender_df
```

#### Exercise 2: Calculate Survival Percentage Grouped by Gender and Class

```python
#count of gender and class based on survival:
gender_class_survival = df2.groupby(['Sex', 'Pclass'])['survived'].sum()
gender_class_survival
```

Due to ambiguity in question in the base denominator to calculate a percentage, the base could either be:
(1) Total passengers in the Passenger Class with that Gender
(2) Total Overall Passengers in whole Titanic

Nevertheless, I will create a dataframe to illustrate both.

```python
total_pax_gender_class = df2.groupby(['Sex', 'Pclass'])['survived'].count()
total_pax_gender_class
```

```python
gender_class_df = pd.DataFrame({
    'Gender Class Survival': gender_class_survival,
    'Total Passengers': total_pax_gender_class
})
gender_class_df
```

```python
gender_class_df.index = gender_class_df.index.set_levels(gender_class_df.index.levels[0].map(gender_labels), level=0)

gender_class_df
```

```python
gender_class_df['Survival Percentage By Gender Class'] = ((gender_class_df['Gender Class Survival'] / gender_class_df['Total Passengers'])*100).round(2)
gender_class_df['Overall Survival Percentage'] = ((gender_class_df['Gender Class Survival'] / len(df2))*100).round(2)

gender_class_df
```
